<h1>Part A</h1>

in recent years the use of text recognition has skyrocketed, this is due to the breakthroughs in artificial intelligence (AI) technology. Optical character recognition (OCR) is the process that converts an image of text into a machine-readable text format (Amazon, 1978). OCR has evolved from being a physical machine “that could only read printed text and convert it into telegram codes” (Hero, 2022) to software that can “recognise coloured fonts or identify text on complex backgrounds.” (Hero, 2022). Its use in contemporary society has been huge and is used in almost all things image related. We use OCR as we move into a paperless society. It takes time and a lot of space to store physical files such as important legal documents, invoices or contracts. Instead, what most people do is scan the documents digitally saving them as images on a server. However, the text in images cannot be used by word processors straight away.  This is where OCR comes in to convert the images into actual text data. OCR works by first scanning the image and groups the light area as a background and the dark area as text. It will then clean the image removing errors so it can get a good read. It will re align the image if it is distorted or rotated in any way and remove any spots or lines and try to clean up as much as it can. It will then start a pattern matching algorithm comparing the image data to the data from the database to see if it matches, it will then convert the extracted data into a file and output that to the user. However, this may not be completely accurate, there are millions of fonts and styles of writing they cannot be all captured and stored in a database, hence why there are different types of OCR. Another common type is Intelligent character recognition (ICR) which is used to read the text the same way humans do. They use far more advanced method to train the machine to think and behave like a human. It uses neural networks so analyse text looking for different attributes, such as curves, lines loops, all of which combine to give a final result. Even though ICR only processes one character at a time, the process is faster, with results coming up in seconds. Furthermore, there is also intelligent word recognition which works the same as ICR, but processes whole word images instead of single characters. And optical mark recognition which identifies things such as logos, watermarks and other symbols in an image. These new advancements in OCR have allowed so many processes in our life to become automated from banking, healthcare and logistics. Banks are able to scan images of bank notes or cheques and use highly advanced neural networks to check if the notes are counterfeit or real Things such as speed cameras are able to take pictures of car number plates when speeding and extract the number plate from the images, automating giving tickets to drivers who are speeding. We also use it for google translate, where people take a picture, or hold their phone up in real time of text that is a different language. OCR scans the image and uses matches the words in the image with googles database, so that they can translate the text for you.

OCR has many techniques, all of which are used to extract text from an image. The first technique used is pattern recognition and feature extraction. “Pattern recognition is defined as a data analysis process that uses machine learning algorithms to classify input data into objects, classes, or categories based on recognized patterns, features, or regularities in data.” (Kanade, 2023) These two methods, combined, are important for OCR because they allow the system to find and recognise the characters. Firstly, the image needs to be prepared. During this stage, the image is scanned, and algorithms and filters are applied to reduce noise and improve the visibility of the image. It then goes onto the next stage, which is text localisation, this is where it identifies the regions where text is present. The programme draws bounding boxes around potential text regions and attempts to use groups of connected pixels to see if they form text. Then it moves on to line and word segmentation, which separates the text into lines and words to help with character-level recognition. The way it does this is by analysing the pixels along rows to identify lines and words. It also identifies the gaps between words to find and separate words. Then it will do character segmentation, which is where each individual character is isolated for individual recognition. This process is similar to line and word segmentation, except this time it analyses the stroke width to locate the characters and separate them from the words. Then the software will begin feature extraction. Once all the characters and words are isolated, it will start to extract the most relevant from the characters to represent the patterns. It will look at a multitude of things to extract, such as edge detection: identifying boundaries and edges; contour analysis: shape and structure of characters; and the Histogram of Oriented Gradients (HOG), which is the shape and gradient information. After this, it will start pattern classification, assigning a label to each isolated feature extracted from the characters. It is also here that the model is trained using the labelled data to learn the patterns associated with the characters. The main strengths of pattern recognition are its versatility, it can recognise patterns in not just only images, but speech, sound and handwriting, meaning it can be used in many different applications. Furthermore, the pattern recognition adapts as it is exposed to more data allowing it to handle more diverse patterns. Another strength is that due to the use of AI and machine learning the algorithms are all automated with little to no human interaction. This eliminates human bias and the need for human interaction speeding the process up. However, there are a few drawbacks one being that the system relies on a large amount of data both to train and validate the model. There also is the danger of overtraining the model, it may memorise specific patterns in the data, instead of creating a generalised representation, this will lead to the model performing well on training data but poorly on new data not seen by the system before. Another weakness could be that because the algorithms are complex, they require significant resources such as computational power and a large storage device to save its data.









The next technique in OCR is Convolutional Neural Networks (CNNs), these are a deep learning architecture specifically designed for image-based tasks. They have three main types of layers these are the convolutional layer, pooling layer and the Fully-connected (FC) layer. The convolutional layer is the core of the CNN and where the majority of the process happens. The process involves a kernel which is a small 2d array, the filter size is typically a 3x3 matrix, this determines the size of the receptive field. This filter moves across the input image checking if a specific feature exists. It will repeat this process scanning the image, after each iteration a dot product is calculated between the input pixels and the filter. The output from the dots is known as a feature map. The image is eventually converted into numerical values in this layer, allowing the CNN to interpret the image and extract the patterns. There are three parameters that effects the size of the output. There are the number of filters, stride and zero padding. The number of filters affects the depth of the output. Having more filters will give more different feature maps. Stride is the distance that the kernel moves over the input. A larger stride yields a smaller output. And zero-padding is usually used when filters so onto fit the input image.it sets all the elements that falls outside the input matrix to 0, producing a larger output. There are also three types of padding: valid padding, same padding and full padding. Valid padding, also known as no padding, is where the last convolution is dropped if the dimensions do not align. Same padding ensures that the output layer has the same size as the input layer. And lastly full padding, increases the size of the output by adding zeros to the border of the input. The pooling layer is similar to the convolution layer, sweeping across the image finding features, but unlike the convolutions layer, the pooling layer reduces the number of parameters in the input and also results in some information loss, this does however reduce complexity and improves the efficiency of the CNN. There are two main types of pooling max pooling and average pooling. Max pooling is when the filter moves, it selects the pixel with the maximum value to send to the output array. Average pooling is while the filer is moving, it calculates the average value within the receptive field and sends it to the output array. Lastly the fully connected layer, this is where each node in the output layer connects directly to a node in the previous layer. This layer classifies features based on the layers and filters, “FC layers usually leverage a softmax activation function to classify inputs appropriately, producing a probability from 0 to 1” (IBM, 2023). The main strength of CNN is that due to it using multiple layers, it can deliver more accurate results. They are also useful because they provide highly accurate results when large amounts of data are involved. The CNN also learns features more after each iteration the object data moves through the layers. This removes the need for any manual feature extraction. Another benefit of CNNs is that they can be retrained for new recognition tasks, so they can be used in multiple different real-world applications without any extra cost, they are also easy to deploy and can run on any device. However, a weakness of CNNs is that it is very data dependant, the large number of algorithms and nodes needs sufficient amount of data to bring an output. Inadequate or biased data can result in poor performance. Due to the vast amounts of nodes and layers, training CNNs can be computationally intensive and resource demanding. Along with the large amount off storage needed to store the data used to train the model.





Another technique used is in OCR is transfer learning. “transfer learning is a machine learning method where we reuse a pre-trained model as the starting point for a model on a new task.”(Baheti, 2021) In transfer learning, you use what has been learned in one task to improve generalisation in another problem by transferring the weights in a network from one network to another. Instead of training an entirely new model with data, you use the knowledge gained from a pretrained model with large amounts of labelled training data in a new task.in normal neural networks, edge detection is done in the early layers, shapes in the middle, and some specific features in the later layers. In transfer layers the early and middle layers are used and the later layers are retrained. An example is a model that is trained to search for a specific word in an image, like the word “hello”, the new model wants to search for the word “world”. In the early layers it will instead search for “world” in the image. It will then try to extract features and patterns related to the new word, and use the later layers to retrain the model to search for the new word. Its mainly used when there isn’t enough labelled training data to train the network from scratch. It is also used when the two tasks have similar input image. There are a few types of transfer learning inductive, transductive and unsupervised. Inductive requires the source and target domains to be the same, the algorithms try to use the knowledge from the source model and apply it to improve the task. Transductive transfer is where the domains of the source and target tasks are similar but not the same. In these situations, there is typically a lot of labelled data in the source domain and very little unlabelled data in the destination domain. And lastly unsupervised, it works similar to inductive, but the main difference is that the algorithms focus on unsupervised tasks and involve unlabelled datasets for both source and target tasks. The steps of transfer learning are as follows: obtaining pre-trained model, creating a base model, freezing layers, adding new trainable layers, training the new layers, fine-tuning the model. In the first step, a pre-trained model is chosen, theses are chosen based on the task. The next step a base model is created, all the weights are downloaded, waving the additional time spent training the model. In step 3 the starting layers are frozen, this is done to ensure that the model does not re learn basic features. If its not frozen it will lose all the learning that had taken place on previous models. In step 4 new layers are added, this is done to add or change specific features or tasks that needs to be done. Step 5 is to train the new layers, the pretrained models’ output will be different to the output we want, so you will have to train the new layers so that it will give the output that you need. The 6th step is about fine tuning the model, unfreezing the model and changing the model slightly to increase the performance of the model. Transfer learning has several benefits, but the main one is saving training time, and not needing a lot of data. It also has improved generalisation, as the knowledge gained from one task is applied to another, this is also helpful when the source and target tasks are related. However, some weaknesses are that the source and target domain have to be similar, if they are too dissimilar transferring knowledge may not be effective and may not give the correct outputs. It may also be difficult to find a source tasks, and choosing a source task that is not optimal may result in poor results.

```python
import pytesseract
from pytesseract import Output
import PIL.Image
import cv2
```

```python
myconfig = r"--psm 6 --oem 3"##changing thr psm value changes the recognition mode
```

```python
text = pytesseract.image_to_string(PIL.Image.open("OCR.png"), config=myconfig)
print(text)
```

```python
text = pytesseract.image_to_string(PIL.Image.open("OCR2.png"), config=myconfig)
print(text)
```

```python
text = pytesseract.image_to_string(PIL.Image.open("OCR3.png"), config=myconfig)
print(text)
```

```python
text = pytesseract.image_to_string(PIL.Image.open("logos.png"), config=myconfig)
print(text)
```

```python
img = cv2.imread("logos.png")
height, width, _ = img.shape
```

```python
##Places boxes around the words, and outputs what word it sees
data = pytesseract.image_to_data(img, config=myconfig, output_type = Output.DICT)

amount_boxes = len(data['text'])
for i in range(amount_boxes):
    if float(data['conf'][i]) > 20:##changing the value changes the confidence kevel of the program allowing it to make bolder decisions
        (x, y, width, height) = (data['left'][i], data['top'][i], data['width'][i], data['height'][i])
        img = cv2.rectangle(img, (x, y), (x + width, y + height), (0, 0, 255), 2)
        img = cv2.putText(img, data['text'][i], (x, y + 80), cv2.FONT_HERSHEY_PLAIN, 2.0, (255,255,255), 2, cv2.LINE_AA)

import matplotlib.pyplot as plt
# load image using cv2....and do processing.
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
# as opencv loads in BGR format by default, we want to show it in RGB.
plt.show()
```
